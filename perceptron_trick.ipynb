```python
from sklearn.datasets import make_classification
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=1, n_classes = 2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, s=100)
plt.show()
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
print(intercept_, coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + b
```

```python
sns.lineplot(x=x_input, y=y_input, color='red')
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, s=100)
plt.ylim(-3, 2)
```
